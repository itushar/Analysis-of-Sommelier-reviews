## Lets find wines in our price point!

```python
import pandas as pd
dataset= pd.read_csv("/home/tsb348/notebooks/winemag-data-130k-v2.csv", index_col=0)
dataset
```

```python
dataset['points_per_dollar']=dataset['points']/dataset['price']
dataset['vintage']=dataset.title.str.extract('(\d+)')
dataset
```

#### College student with max budget of 10 dollar per bottle, wants to be mildly sophisticated with a min desired price of 5 dollar per bottle and wants list of 5 Pinot Noir bottles in price range that have most bang for the buck

```python
def curate_list_student (max_price, min_price, wine_type):
    wine_list = (dataset.loc[(dataset['price'] <= max_price) & (dataset['price'] >= min_price) & (dataset['variety'] == wine_type)])
    wine_list_sort = wine_list.sort_values(by = ['points_per_dollar'], ascending=False)
    wine_list_sort.drop_duplicates(subset='title', keep='first', inplace=True)
    return wine_list_sort.head(5)

curate_list_student(10, 5, 'Pinot Noir')
```

#### Fancy party host wants to impress guests with highest rated wine from France within price range, lower price is tie-breaker for wines with the same rating.


```python

def curate_list_host (max_price, country):
    wine_list = (dataset.loc[(dataset['price'] <= max_price) & (dataset['country'] == country)])
    wine_list_sort = wine_list.sort_values(['points','price'], ascending=[0,1])
    wine_list_sort.drop_duplicates(subset='title', keep='first', inplace=True)
    return wine_list_sort.head(5)

curate_list_host(100, 'France')
```

#### Collector wants to know best wines of a certain type and vintage from a particular state or province, lower price is tie-breaker for wines with the same rating


```python

def curate_list_collector (province, wine_type, vintage):
    wine_list = (dataset.loc[(dataset['province'] == province) & (dataset['variety'] == wine_type) & (dataset['vintage'] == vintage)])
    wine_list_sort = wine_list.sort_values(['points','price'], ascending=[0,1], na_position='last')
    wine_list_sort.drop_duplicates(subset='title', keep='first', inplace=True)
    return wine_list_sort.head(5)

curate_list_collector('Bordeaux', 'Bordeaux-style White Blend', '2013')
```

#### Jaccard Similarity is the simplest of the similarities and is nothing more than a combination of binary operations of set algebra.

#### To calculate the Jaccard Distance or similarity we treat the description as a set of tokens. Mathematically the formula is as follows:

####             Jaccard Similarity(A,B) = Intersection(A,B)/ Union(A,B)

```python
import string
import math

def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)

#Running for first 10 only
no_price = no_price.head(10).copy()
no_price['jaccard_distance'] = no_price.apply(lambda _: float(0), axis=1).copy()


for i in no_price.index:
    v_price = with_price[with_price.variety == (no_price['variety'][i])].copy()
    for j in v_price.index:
        checl = jaccard_similarity(no_price['tokenized_description'][i], v_price['tokenized_description'][j])
        temp = max(no_price['jaccard_distance'][i], checl)
        no_price.at[i,'jaccard_distance'] = temp
        no_price.at[i,'closest_match'] = j
        no_price.at[i,'price_estimate'] = v_price['price'][j]
```
